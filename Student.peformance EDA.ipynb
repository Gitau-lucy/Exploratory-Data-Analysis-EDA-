```python
import matplotlib.pyplot as plt

# Histogram
df['age'].hist()
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart (categorical)
df['sex'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
df[['G1', 'G2', 'G3']].corr()
```

```python
import matplotlib.pyplot as plt

plt.imshow(df[['G1','G2','G3']].corr())
plt.colorbar()
plt.xticks(range(3), ['G1','G2','G3'])
plt.yticks(range(3), ['G1','G2','G3'])
plt.title("Correlation Between Grades")
plt.show()
```

```python
df.groupby('studytime')['G3'].mean()
```

```python
df.groupby('studytime')['G3'].mean().plot(kind='bar')
plt.title("Average Final Grade by Study Time")
plt.xlabel("Study Time Level")
plt.ylabel("Average G3")
plt.show()
```
